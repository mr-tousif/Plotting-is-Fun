```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive
```

```python
x=np.linspace(0,20,100)
y=np.linspace(0,20,100)
```

```python
x,y=np.meshgrid(x,y)

def f(A,B):
    return np.sin(np.sqrt(A**2+B**2))

z=f(x,y)

ax=sns.heatmap(z)

#ax.invert_yaxis() to get the perfect ordering of values 

ax.invert_yaxis()
```

```python
x=np.linspace(0,20,100)
y=np.linspace(0,20,100)

x,y=np.meshgrid(x,y)

def f(A,B,C,D):
    return D*np.sin(C*np.sqrt(A**2+B**2))

def plotting(C=1,D=1):
    z=f(x,y,C,D)
    ax=sns.heatmap(z)  
    ax.invert_yaxis()  #ax.invert_yaxis() to get the perfect ordering of values 
    
interactive_plot=interactive(plotting, C=(0,10,1),D=(1,5,1))

interactive_plot
```

```python
x=np.linspace(0,20,100)
y=np.linspace(0,20,100)

x,y=np.meshgrid(x,y)

def f(A,B,C,D):
    return D*np.sin(C*np.sqrt(A**2+B**2))

def plotting(C=1,D=1):
    z=f(x,y,C,D)
    ax=sns.heatmap(z, vmin=-5,vmax=5)  
    ax.invert_yaxis()  #ax.invert_yaxis() to get the perfect ordering of values 
    
interactive_plot=interactive(plotting, C=(0,10,1),D=(1,5,1))

interactive_plot
```

```python
x=np.linspace(0,20,100)
y=np.linspace(0,20,100)

x,y=np.meshgrid(x,y)

def f(A,B,C,D):
    return D*np.sin(C*np.sqrt(A**2+B**2))

def plotting(C=1,D=1):
    z=f(x,y,C,D)
    ax=sns.heatmap(z,xticklabels=10,yticklabels=10, vmin=-5,vmax=5)  #giving separation 10 to x and y axis
    ax.invert_yaxis()  #ax.invert_yaxis() to get the perfect ordering of values 
    plt.title("Introduction to Heatmap")
    
interactive_plot=interactive(plotting, C=(0,10,1),D=(1,5,1))

interactive_plot
```

Meshgrid

```python
x_1=np.linspace(-2,2,100)
x_2=np.linspace(-1,1,100)

x_11,x_22=np.meshgrid(x_1,x_2)
f=np.exp(-x_11**2-x_22**2)

plt.figure(figsize=(8,4))
plt.pcolormesh(x_11,x_22,f)
plt.colorbar()
plt.show()
```

Creating a rectangular mask:

\begin{equation} \label{eq1}
\begin{split}
A & ={x^2+y^2<1} \\
 & = {x^2+y^2\geq1}   
\end{split}
\end{equation}
      

```python
x=y=np.linspace(-5,5,100)


x_1,y_1=np.meshgrid(x,y)
f=(x_1**2+y_1**2<1).astype(float)

plt.figure(figsize=(8,4))
plt.pcolormesh(x_1,y_1,f)
plt.colorbar()
plt.show()
```

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm  #colormap




x=np.linspace(1,10,100)
y=np.linspace(1,10,100)

xv,yv=np.meshgrid(x,y)
f=lambda x,y: np.sin(x)**2+np.cos(y)
z=f(xv,yv)
#plt.contour(xv,yv,z,50)


fig=plt.figure(figsize=(10,5))

ax=fig.gca(projection='3d')  #to show the axis as a 3D plot
ax.plot_surface(xv,yv,z,cmap=cm.rainbow)
plt.show()
```

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive
from mpl_toolkits.mplot3d import Axes3D
#from matplotlib import cm  #colormap




x=np.linspace(1,10,100)
y=np.linspace(1,10,100)

xv,yv=np.meshgrid(x,y)
f=lambda x,y: np.sin(x)**2+np.cos(y)
z=f(xv,yv)
#plt.contour(xv,yv,z,50)


def plotter(E,A):   #interactive viewing angle plot
    fig=plt.figure(figsize=(10,5))
    ax=fig.gca(projection='3d')  #to show the axis as a 3D plot
    ax.plot_surface(xv,yv,z,cmap="coolwarm")
    ax.view_init(elev=E, azim=A)
    plt.show()
    
plotter(45,45)
```

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interactive
from mpl_toolkits.mplot3d import Axes3D
#from matplotlib import cm  #colormap




x=np.linspace(1,10,100)
y=np.linspace(1,10,100)

xv,yv=np.meshgrid(x,y)
f=lambda x,y: np.sin(x)**2+np.cos(y)
z=f(xv,yv)
#plt.contour(xv,yv,z,50)


def plotter(E,A):   #interactive viewing angle plot
    fig=plt.figure(figsize=(10,5))
    ax=fig.gca(projection='3d')  #to show the axis as a 3D plot
    ax.plot_surface(xv,yv,z,cmap="coolwarm")
    ax.view_init(elev=E, azim=A)
    plt.show()
    
#plotter(45,45)

interactive_plot=interactive(plotter, E=(-90,90,4), A=(-90,90,4))
interactive_plot
```
